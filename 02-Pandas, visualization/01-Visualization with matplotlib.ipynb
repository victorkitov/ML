## Displaying several functions

### On one graph

```python
x=linspace(0,3,30)
plot(x,x**2,color='r')
plot(x,x**3,color='b')
```

### Labelling

```python
plot(x,x**2,color='r', label='y=$x^2$')
plot(x,x**3,color='b', label='y=$x^3$')
legend(loc='best')
title('Comparison of functions')
xlabel('x')
ylabel('f(x)')
```

### Different markers

```python
plot(x,x**2,color='r')
plot(x,x**3,color='b')
plot(x,20*sin(x),color='r',linestyle=':')
plot(x,20*cos(x),color='b',marker='x',lw=0)
```

### Visualization on several graphs

```python
figure()
plot(x,x**2)

figure()
plot(x,x**3)
```

```python
subplot(2,1,1)  # 2 plots along verical axis and one plot along horizontal axis, select 1st
title('1st plot')
plot(x,x**2)
subplot(2,1,2)  # 2 plots along verical axis and one plot along horizontal axis, select 2nd
title('2nd plot')
plot(x,x**3)
```

```python
subplot(1,2,1)  # 2 plots along verical axis and one plot along horizontal axis, select 1st
title('1st plot')
plot(x,x**2)
subplot(1,2,2)  # 2 plots along verical axis and one plot along horizontal axis, select 2nd
title('2nd plot')
plot(x,x**3)
```

## Visualize $(x_1,x_2,y)$, $x_1, x_2$ - real, $y$ - categorical

```python
n=100
X=vstack([randn(n,2),randn(n,2)+3])
Y=hstack([zeros(n),ones(n)])
```

```python
colors={0:'b', 1:'r'}
scatter(X[:,0], X[:,1], c=[colors[y] for y in Y], lw=0)
```

## Visualize $(x_1,x_2,y,z)$,$y$-categorical,  $x_1,x_2,z$-real

```python
n=100
X=vstack([randn(n,2),randn(n,2)+3])
Y=hstack([zeros(n),ones(n)])
Z=rand(200)**2
```

```python
colors={0:'b', 1:'r'}
scatter(X[:,0], X[:,1], c=[colors[y] for y in Y], s=5+30*Z, edgecolor='k')
```

## Visualize $(x_1,x_2,y,z)$, $x_1, x_2$ - real,  $y, z$ - categorical

```python
n=100
X=vstack([randn(n,2),randn(n,2)+3])
Y=hstack([zeros(n),ones(n)])
Z=rand(200)>0.2
```

```python
colors={0:'b', 1:'r'}
markers={0:'o', 1:'x'}
for z in unique(Z):
    XX = X[Z==z]
    YY = Y[Z==z]
    scatter(XX[:,0], XX[:,1], c=[colors[y] for y in YY], lw=1, s=30, marker=markers[z], edgecolor='k')
```

## Compare $K$ real valued outcomes

Plot bar for each outcome. Bar plot gives comparative analysis.

```python
X=arange(10)
Y=X**2
bar(X,Y)
```

```python
barh(X,Y)
```

```python
colors={True:'r',False:'b'}
barh(X,Y,color=[colors[y>50] for y in Y])
```

```python
val = 3+10*rand(5)    # the bar lengths
pos = arange(5)+.5    # the bar centers on the y axis

figure(1)
barh(pos,val, align='center')
yticks(pos, ('alternative method', 'method 1', 'method 2', 'method 3', 'method 4'))
xlabel('Accuracy')
title('Performance of methods')
grid(True)
```

Add information about standard deviations to show statistical significance.

```python
figure(2)
barh(pos,val, xerr=rand(5), ecolor='r', align='center')
yticks(pos, ('alternative method', 'method 1', 'method 2', 'method 3', 'method 4'))
xlabel('Performance')
```

### Comparative bar plot

```python
val1 = 3+10*rand(5)    # the bar lengths
val2 = 2+10*rand(5)    # the bar lengths
pos = arange(5)    # the bar centers on the y axis
h=0.3

before = barh(pos,val1, height=h, color='b')
after = barh(pos+h, val2, height=h, color='r')
yticks(pos, ('method 1', 'method 2', 'method 3', 'method 4', 'method 5'))
xlabel('Accuracy')
title('Performance of methods')

legend( (before, after), ('before modification', 'after modification') )

grid(True)
```

## Stacked bars

```python
N = 5
Means1 = (18, 20, 24, 30, 45)
Means2 = (25, 32, 34, 20, 25)
Std1 = (2, 3, 4, 4, 2)
Std2 = (3, 5, 2, 3, 3)
ind = np.arange(N)    # the x locations for the groups
width = 0.35       # the width of the bars: can also be len(x) sequence

p1 = bar(ind, Means1, width, color='r', yerr=Std1)
p2 = bar(ind, Means2, width, color='y',
             bottom=Means1, yerr=Std2)

ylabel('Total earnings')
title('Total earnings by product and year')
xticks(ind + width/2., ('2016', '2017', '2018', '2019', '2020'))
yticks(np.arange(0, 81, 10))
legend((p1[0], p2[0]), ('Product 1', 'Product 2'))
```

# Show function of 2 arguments

```python
def fun(x, y):
    return x**2+(y-3)**2
```

```python
n = 256
X = np.linspace(-10, 10, n)
Y = np.linspace(-6, 6, n)
#XX, YY = np.meshgrid(x, y)


F = zeros([n,n])
for (ix,x) in enumerate(X):
    for (iy,y) in enumerate(Y):
        F[iy,ix] = fun(x,y)  # pay attention to order: iy, ix
        
plt.contourf(X, Y, F, 18, c='black', alpha=.75, cmap='hot')
colorbar()
plt.contour(X, Y, F, 18, colors='black', linewidth=.5)
xlabel('x')
ylabel('y')
```

# 3D plots

```python
from mpl_toolkits.mplot3d import Axes3D
```

```python
fig = plt.figure()
ax = Axes3D(fig)
X = np.arange(-4, 4, 0.25)
Y = np.arange(-4, 4, 0.25)
X, Y = np.meshgrid(X, Y)
Z = X**2+Y**2

ax.plot_surface(X, Y, Z, rstride=1, cstride=1, cmap='hot')
```

```python
fig = plt.figure()
ax = Axes3D(fig)
X = randn(50)
Y = randn(50)
Z = randn(50)

ax.scatter(X, Y, Z)
```
